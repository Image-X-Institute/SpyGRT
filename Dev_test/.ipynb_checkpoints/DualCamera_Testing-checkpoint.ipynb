```python
from spygrt.stream import Camera, DualCamera
from spygrt.tracker import Tracker
from spygrt.calibration import Calibrator
import open3d as o3d
import numpy as np
import pyrealsense2 as rs2
import time
import matplotlib.pyplot as plt
```

```python
cam = []
ctx = rs2.context()
a1=0
b1=0
for dev in ctx.query_devices():
    a1 = time.time()
    cam.append(Camera(dev))
    b1 = time.time()
    print(a1-b1)
```

```python
d = DualCamera(cam[0],cam[1])
```

```python
d.start_stream()
```

```python
d.get_frames('o3d')
```

```python
d.timestamp
```

```python
d.compute_pcd()
```

```python
d.pose
```

```python
calibs = (Calibrator(d.stream1), Calibrator(d.stream2))
```

```python
for i in range (100):
    calibs[0].find_corners3d(d.stream1.get_frames('rs'))
    calibs[1].find_corners3d(d.stream2.get_frames('rs'))
```

```python
calibs[0].corners_scale()
calibs[1].corners_scale()
calibs[0].align_cameras(calibs[1])
calibs[1].align_to_board(cal2 = calibs[0])
```

```python
for cal in calibs:
    cal.write_cal()
```

```python
d.load_calibration()
```

```python
d.stream1.load_calibration()
d.stream2.load_calibration()
```

```python
pcd = d.compute_pcd()
```

```python
o3d.visualization.draw_geometries([pcd.to_legacy()])
```

```python
a1=0
b1=0
a2=0
b2=0
for i in range(100):
    a1 += time.time()
    d.stream1.get_frames('o3d')
    d.stream2.get_frames('o3d')
    b1 += time.time()
print(b1-a1)
```

```python
ext = o3d.core.Tensor(np.identity(4), dtype=o3d.core.Dtype.Float32, device=d.stream1.pcd.device)
ext[2][3] = 0.75
ext[1][1] = -1
rgbd = d.stream1.compute_pcd(True).project_to_rgbd_image(1280,720,d.stream1.intrinsics,ext)
plt.imshow(rgbd.to_legacy().color)
plt.show()
```
